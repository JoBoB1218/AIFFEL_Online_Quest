```python
# MainQeust01-1
# 조보겸 / 2023-11-17 / 은행 계좌 문제

import random

class Account:

    # 계좌 개수를 출력하는 클래스 변수
    num_account = 0

    # 생성자
    def __init__(self, account_name, account_stack):
        self.account_bank = 'SC은행'
        self.account_name = account_name
        self.account_stack = account_stack
        self.account_number = f'{random.randint(0,999):0>3}-{random.randint(0,99):0>2}-{random.randint(0,999999):0>6}'

        self.deposit_count = 0
        self.print_depoist_history = ''
        self.print_withdraw_history = ''

        Account.num_account +=1

    # 계좌 개수를 출력해주는 메소드
    def get_account_num(self):
        return Account.num_account

    # 최소 1원 이상만 입금이 가능한 메소드
    def deposit(self, cash):

        # 입금
        if cash >= 1:
            self.account_stack += cash
            print('현재 돈 :',self.account_stack)

            # 입금 내용 기록
            self.print_depoist_history += f'입금 : {cash} \n'

            # 입금 횟수 1회 증가
            self.deposit_count += 1

        # 입금 실패
        else:
            print('1원 이하는 입금 불가입니다!')

        # 만약 입금 횟수가 5배수라면 이자 추가
        if self.deposit_count % 5 == 0:

            # 돈에 소수점 자리는 없기 때문에 소수점으로 들어가는 값은 버렸다.
            self.account_stack += int(self.account_stack*1/100)
            print('이자가 붙었습니다!')
            print('현재 돈 : ', self.account_stack)

            # 이자 기록
            self.print_depoist_history += f'이자 : {int(self.account_stack*1/100)} \n'



    # 잔고 이하만 출금 가능한 메소드
    def withdraw(self, cashback):

        # 출금
        if cashback <= self.account_stack:
            self.account_stack -= cashback
            print('현재 돈 :',self.account_stack)

            # 출금 내역 기록
            self.print_withdraw_history += f'출금 : {cashback} \n'

        # 출금 실패
        else:
            print('계좌 잔고 이상으로는 출금하실 수 없습니다!')

    # 계좌 정보들을 출력해주는 메소드
    def display_info(self):
        cash = format( self.account_stack, ',')
        print(f'은행이름 : {self.account_bank}, 예금주 : {self.account_name}, 계좌번호 : {self.account_number}, 잔고 : {cash}')

    # 입금 내역을 출력해주는 메소드
    def deposit_history(self):
        print(self.print_depoist_history)

    # 출금 내역을 출력해주는 메소드
    def withdraw_history(self):
        print(self.print_withdraw_history)



# 예금주명과 초기 잔액 입력
account1 = Account(account_name = input('예금주명 :'), account_stack = int(input('초기 잔액 :')))
account2 = Account(account_name = input('예금주명 :'), account_stack = int(input('초기 잔액 :')))
account3 = Account(account_name = input('예금주명 :'), account_stack = int(input('초기 잔액 :')))


# 계좌들의 list
account_list = []

# 계좌 list 추가
account_list.append(account1)
account_list.append(account2)
account_list.append(account3)

# 백만원 이상의 잔고를 가진 고객의 정보 출력
for account in account_list:
    if account.account_stack >= 1000000:
        account.display_info()

print()

# 입금 5회 진행
for i in range(5):
    account1.deposit(cash = int(input('입금하실 금액을 입력해주세요 : ')))
    print()

# 출금 5회 진행
for i in range(5):
    account1.withdraw(cashback = int(input('출금하실 금액을 입력해주세요 : ')))
    print()

print()

# 입금 내역 출력
account1.deposit_history()

print()

# 출금 내역 출력
account1.withdraw_history()
```
